```python

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the data from the text file
filename = "/content/drive/MyDrive/Education/Phd/research/SLM/Data Visualization/Macmini/Phi-2/disk_busy_rate.txt"

# Initialize lists to hold the data
data = {
    "disk0": [],
    "disk10": [],
    "disk12": [],
    "disk14": []
}

with open(filename, 'r') as file:
    lines = file.readlines()

    # Debug: Print the header
    print("Header:", lines[0])

    for line in lines[1:]:  # Skip the header
        if line.strip():
            parts = line.split()
            # Debug: Print the line parts
            print("Line parts:", parts)
            if len(parts) >= 12:  # Ensure line has the correct number of parts
                try:
                    data["disk0"].append(int(parts[1]))
                    data["disk10"].append(int(parts[4]))
                    data["disk12"].append(int(parts[7]))
                    data["disk14"].append(int(parts[10]))
                except ValueError as e:
                    # Debug: Print the error message
                    print("Error parsing line:", line)
                    print(e)

# Step 2: Convert to pandas DataFrame
data = pd.DataFrame(data)

# Debug: Print the DataFrame to check the parsed data
print(data)


# Step 3: Calculate the combined tps
m_p_db['Disk Usage %'] = data.sum(axis=1)

#m_p_db['Disk Usage %'] = (data['combined_tps'] / data['combined_tps'].max()) * 100


# Step 4: Visualize the combined tps
plt.figure(figsize=(10, 6))
plt.plot(data['combined_tps'], label='Combined TPS')
plt.plot(m_p_db['Disk Usage %'], label='Disk Usage %')
plt.xlabel('Time Intervals')
plt.ylabel('TPS')
plt.title('Combined TPS Over Time')
plt.legend()
plt.grid(True)
plt.show()





```
