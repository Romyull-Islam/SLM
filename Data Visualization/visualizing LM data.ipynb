```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
cpu_usage = pd.read_csv("/content/drive/MyDrive/Education/Phd/research/SLM/Data Visualization/Tinyllama_csv/CPU_ALL.csv")
cpu_usage.head()
```

```python
# prompt: visualize all the datas of the csv in one time series graph. here CPU Total UbuntuRomy is time

# Set the time series as the index
cpu_usage.set_index("CPU Total UbuntuRomy", inplace=True)

# Plot all the data in the CSV file as a time series graph
cpu_usage.plot(figsize=(12, 6))

# Add a title and labels to the axes
plt.title("Time Series Graph of All Data in CPU_ALL.csv")
plt.xlabel("Time (CPU Total UbuntuRomy)")
plt.ylabel("Value")

# Show the plot
plt.show()
```
